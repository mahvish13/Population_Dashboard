```python
df["Urban Pop %"] = pd.to_numeric(
    df["Urban Pop %"].str.replace("%", ""),
    errors="coerce"
)

df["World Share"] = pd.to_numeric(
    df["World Share"].str.replace("%", ""),
    errors="coerce"
)
```

```python
df["Urban Pop %"] = df["Urban Pop %"].fillna(df["Urban Pop %"].median())
df["World Share"] = df["World Share"].fillna(df["World Share"].median())
```

```python
population=df["Population (2020)"].to_numpy()
mean=np.mean(population);
print(mean)
meadian=np.median(population)
print(meadian)
std=np.std(population)
print(std)
```

```python
density = df["Density (P/KmÂ²)"].to_numpy()

percentiles = np.percentile(density, [25, 50, 75])

print("25th Percentile:", percentiles[0])
print("50th Percentile (Median):", percentiles[1])
print("75th Percentile:", percentiles[2])
```

```python
labels=np.where(density>500,"High Density" ,"Low Density")
print(labels)
```

```python
log_population=np.log(population)
print(log_population)
```

```python
df["Fert. Rate"] = pd.to_numeric(
    df["Fert. Rate"],
    errors="coerce"
)

df["Med. Age"] = pd.to_numeric(
    df["Med. Age"],
    errors="coerce"
)
```

```python
df["Fert. Rate"] = df["Fert. Rate"].fillna(df["Fert. Rate"].median())
df["Med. Age"] = df["Med. Age"].fillna(df["Med. Age"].median())
```

```python
fert_rate=df["Fert. Rate"].to_numpy()
med_age=df["Med. Age"].to_numpy()
correlation=np.corrcoef(fert_rate , med_age)
print(correlation)
```

```python
top10=df.nlargest(10,"Population (2020)")
print(top10[["Country (or dependency)","Population (2020)"]])
bottom10=df.nsmallest(10,"Population (2020)")
print(bottom10[["Country (or dependency)","Population (2020)"]])
```

```python

sorted_df = df.sort_values(by="World Share", ascending=False)

sorted_df["Cumulative Share"] = sorted_df["World Share"].cumsum()

result = sorted_df[sorted_df["Cumulative Share"] <= 50]

print(result)

print("Number of countries:", len(result))
```

```python
negative = df[df["Migrants (net)"] < 0]

positive = df[df["Migrants (net)"] > 0]

print("Net Emigration:", negative.shape[0])

print("Net Immigration:", positive.shape[0])
```

```python
corr_matrix = df.corr(numeric_only=True)

print(corr_matrix)
```

```python
df["Population Density Category"] = pd.cut(
    df["Density (P/KmÂ²)"],
    bins=[0, 100, 500, 1000, float("inf")],
    labels=["Low", "Medium", "High", "Very High"]
)

print(df[["Density (P/KmÂ²)", "Population Density Category"]].head())
```

```python
highest_fert = df.sort_values(by="Fert. Rate", ascending=False).head(1)

highest_age = df.sort_values(by="Med. Age", ascending=False).head(1)

print(highest_fert[["Country (or dependency)", "Fert. Rate"]])

print(highest_age[["Country (or dependency)", "Med. Age"]])
```

```python
result = df[df["Urban Pop %"] > 90]

result = result.sort_values(by="Population (2020)", ascending=False)

print(result[["Country (or dependency)", "Urban Pop %", "Population (2020)"]])
```
